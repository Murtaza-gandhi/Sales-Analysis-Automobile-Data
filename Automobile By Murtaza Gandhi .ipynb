# Requirement 1

# what is the average price and average mileage for convertible cars

```python
##what is the average price and average mileage for convertible cars

avg_price_milage = pd.DataFrame(df.groupby('body-style')['price','average-mileage'].mean().reset_index()).head(1)
avg_price_milage.rename(columns={"price":"Avg_price"},inplace=True)
avg_price_milage
```

# Requirement 2

# Sort cars by company and price in decending order

```python
company_price = df.sort_values(["company","price"],ascending=False )
company_price
```

# Requirement 3 

```python

Toyota = df[df['company']=='toyota']
Toyota 
```

# Requirement 4

# which company is producing most expensive car?

```python

exp_car = df[['company','price']].sort_values('price').head(1) 
exp_car
```

# Requirement 6

# company wise average price of cars and order them in decending order

```python
ordering_by_avg = df.copy()
ordering_by_avg['Avg-price'] = df['price'].mean()
ordering_by_avg.sort_values(["company","Avg-price"],ascending=False)
ordering_by_avg.insert(2,"Average-price",ordering_by_avg["Avg-price"])
```

```python
ordering_by_avg.drop("Avg-price",axis=1,inplace=True)
```

```python
ordering_by_avg  
```

# Requirement 7

# create a column named: "decision", If wheel-base is less than 100 and length is less than 175 then "don't buy" else "buy"


```python
df["decision"] = np.where((df['wheel-base']<=100) & (df['length']<=175),"Don't Buy","Buy")
```

# Requirement 8 

# Select only cars whose decision is "buy"; find average price based on company of car

```python
decision  =  df[df['decision']=="Buy"]
```

```python
writer = pd.ExcelWriter("C:\\Users\\Murtaza G\\Desktop\\Python\\Data Sets\\Automobile\\Automobile Analysis.xlsx")

avg_price_milage.to_excel(writer,sheet_name = "Avg Price and Milage ",index=False)

company_price.to_excel(writer,sheet_name = "Price by Company",index=False)

Toyota.to_excel(writer,sheet_name = "Toyota",index=False)

exp_car.to_excel(writer,sheet_name = "Most Expancive Car",index=False)

carwith_ML.to_excel(writer,sheet_name = "Max l-wb-hp Cars",index=False)

carwith_HB.to_excel(writer,sheet_name = "Max l-wb-hp Cars",startcol=0,startrow=3,index=False)

carwith_MH.to_excel(writer,sheet_name = "Max l-wb-hp Cars",startrow=9,startcol=0,index=False)

ordering_by_avg.to_excel(writer,sheet_name = "Company wise Avg",index=False)

decision.to_excel(writer,sheet_name = "Good Cars",index=False)

writer.save()
```
